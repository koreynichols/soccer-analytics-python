```python
import pandas as pd
from statsbombpy import sb

df = sb.events(match_id=3869685)
df.columns
```

```python
df = df[df['type'] == 'Pass']
```

```python
df[['x','y']] = df['location'].apply(pd.Series)
```

```python
df['pass_outcome'].head(50)
```

```python
df = df[df['pass_outcome'].isna()]
```

```python
from mplsoccer import Pitch

pitch = Pitch(pitch_type='statsbomb', pitch_color='white', line_color='black', line_zorder=2)
fig, ax = pitch.draw()

df_arg = df[df['team_id'] == 779]

kde = pitch.kdeplot(
    df_arg.x,
    df_arg.y,
    fill = True,
    thresh=0,
    n_levels=100,
    cut=10,
    ax=ax,
    cmap = 'plasma'
)

ax.set_title('Argentina Starting Sucessfull Pass Locations vs France')
```
